```python
# Create a text file and write in it
f = open("05.txt", "w")
f.write("Ahmed")
f.close()
```

```python
# To read it 
with open("05.txt") as txt:
    for f in txt:
        print(f)
```

```python
# Or like that 
f = open("05.txt", "r")
for a in f:
    print(a)
```

```python
# To append 
f = open("05.txt", "a")
f.write(" \n Alaa")
with open("05.txt") as txt:
    for i in txt:
        print(i)
```

```python
# For Dealing with arabic text you must first make the encoding with "utf8"
f = open("7.txt", "w", encoding="utf8")
f.write("اهلا ومرحبا بكم")
f.close()
```

```python
# To append
f = open("7.txt", "a", encoding="utf8")
f.write("\nالسطر الثاني")
f.write("\nالسطر الثالث")
f.write("\nالسطر الرابع")
f.write("\nالسطر الخامس")
f.close()
```
